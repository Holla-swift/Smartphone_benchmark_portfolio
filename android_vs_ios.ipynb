```python
### best performing smartphones

phone_cpu = all_data.nlargest(10, 'CPU')
phone_cpu
```

```python
### Charts
import matplotlib.pyplot as plt

plt.bar(phone_cpu.Device, phone_cpu.CPU)
plt.xticks(rotation='vertical', size=10)
plt.title('Smartphone CPU performance score', size=14)
plt.ylabel('CPU performance score')
plt.xlabel('Phone specs')

plt.show()
```

```python
### poorest performing smartphones

phone_cpu2 = all_data.nsmallest(10, 'CPU')
```

```python
### Charts

plt.bar(phone_cpu2.Device, phone_cpu2.CPU)
plt.xticks(rotation='vertical', size=10)
plt.title('Smartphone CPU performance score', size=14)
plt.ylabel('CPU performance score')
plt.xlabel('Phone specs')

plt.show()
```

```python
### best UX smartphones

phone_ux = all_data.nlargest(10, 'UX')
```

```python
### Charts

plt.bar(phone_ux.Device, phone_ux.UX, color='r')
plt.plot(phone_ux.Device, phone_ux.MEM, color='blue')

plt.xticks(rotation='vertical', size=10)
plt.title('Smartphones with best UX and MEM performance score', size=14)
plt.ylabel('performance score')
plt.xlabel('Phone specs')
plt.legend(['UX', 'MEM'])

plt.show()
```

```python
### Overall performance

overall = all_data.nlargest(20, 'Total_score')
overall
```

```python
### Charts

plt.bar(overall.Device, overall.Total_score/10**6, color='r')

plt.xticks(rotation='vertical', size=10)
plt.title('Smartphones overall performance score', size=14)
plt.ylabel('performance score in millions')
plt.xlabel('Phone specs')


plt.show()
```
